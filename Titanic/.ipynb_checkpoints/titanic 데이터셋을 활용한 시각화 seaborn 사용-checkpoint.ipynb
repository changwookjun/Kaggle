```python
import pandas as pd
```

```python
train = pd.read_csv('./input/train.csv', index_col='PassengerId')
print(train.shape)
train.head(2)
```

```python
train.columns
```

 Seaborn 사용
=============
1. countplot (x,y 축 중에 하나만 선택 가능, classification)
2. barplot (x, y 축 둘 다 사용 가능, regression)
3. pointplot
4. distplot(histogram 분포도)
5. implot(scatter plot 산점도)

```python
%matplotlib inline
import seaborn as sns
```

```python
sns.countplot(data=train, x='Embarked')
```

```python
sns.countplot(data=train, x='Embarked', hue='Survived')
```

```python
sns.countplot(data=train, x='Pclass', hue='Survived')
```

2. barplot
==========

```python
sns.barplot(data=train, x='Pclass', y='Fare', hue="Survived")
```

3.pointplot
============
x축의 값이 서로 연관성이 있을 경우에는 pointplot

x축의 값이 서로 연관성이 없을 경우에는 barplot

barplot은 연관관계 잘 보이지 않으므로, 연관관계를 확인 하려면 pointplot를 사용해라

```python
sns.pointplot(data=train, x='Pclass', y='Fare', hue="Survived")
```

4. distplot
===========

```python
# null 값이 많아서 오류가 발생 함.
sns.distplot(train['Age'], hist=False)
```

```python
#Age 컬럼의 null 값을 처리
mean_age = train['Age'].mean()
mean_age
```

```python
train.loc[train['Age'].isnull(),'Age'] = mean_age
```

```python
sns.distplot(train['Age'], hist=True)
```

```python
sns.distplot(train['Fare'])
```

```python
# low_fare 를 생성
low_fare = train[train['Fare']<100]
low_fare
```

```python
# low_fare (fare가 100보다 작은)DataFrame을 생성
sns.distplot(low_fare['Fare'])
```

```python
#  낮은 요금을 지불할 승객은 죽을 확률이 매우 높았다.
perish = low_fare[low_fare['Survived'] ==0]
survived = low_fare[low_fare['Survived'] == 1]
```

```python
type(perish)
```

```python
sns.distplot(perish['Fare'], hist=False, label="Perish")
sns.distplot(survived['Fare'], hist=False, label="Survived")
```

5. implot(scatter plot 산점도)
============================

```python
sns.lmplot(data=train, x="Age", y="Fare", hue="Survived")
```

```python
sns.lmplot(data=train, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

```python
# Fare가 500 이상인 데이터를 outlier(이상치)를 제거
low_fare = train[train['Fare'] < 500]
sns.lmplot(data=low_fare, x='Age', y='Fare', hue='Survived', fit_reg=False)
```

```python
# Fare가 100 이상인 데이터를 outlier(이상치)를 제거
low_fare = train[train['Fare'] < 100]
sns.lmplot(data=low_fare, x='Age', y='Fare', hue='Survived', fit_reg=False)
```

6. Subplots(한개의 figure에 여러개의 plot 그리기)
===========================================

```python
import matplotlib.pyplot as plt
```

```python
figure, (ax1, ax2, ax3)= plt.subplots(nrows=1, ncols=3)
figure.set_size_inches(18,4)
sns.countplot(data=train, x='Pclass', hue="Survived", ax=ax1)
sns.countplot(data=train, x='Sex', hue="Survived", ax=ax2)
sns.countplot(data=train, x='Embarked', hue="Survived", ax=ax3)
```

```python
figure, ((ax1, ax2, ax3),(ax4, ax5, ax6))= plt.subplots(nrows=2, ncols=3)
figure.set_size_inches(18,8)

sns.countplot(data=train, x='Pclass', hue="Survived", ax=ax1)
sns.countplot(data=train, x='Sex', hue="Survived", ax=ax2)
sns.countplot(data=train, x='Embarked', hue="Survived", ax=ax3)

sns.countplot(data=train, x='SibSp', hue="Survived", ax=ax4)
sns.countplot(data=train, x='Parch', hue="Survived", ax=ax5)
sns.countplot(data=train, x='Parch', hue="Survived", ax=ax6)
```

```python
figure, ax1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18,4)
sns.countplot(data=train, x="Age", hue="Survived", ax=ax1)
```
